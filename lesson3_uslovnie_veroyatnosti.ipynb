Задание 1

Из 500 компьютеров  180 принадлежат к 1 партии, 170 – ко второй партии, остальные к третьей. В первой партии 3% брака, во второй — 2%, в третьей – 6%. Случайно выбирается один компьютер. Определить вероятность того, что выбранный компьютер – бракованный.  Округлите полученный ответ до десятитысячных. В качестве разделителя используйте точку.

```python
#задаем количества компьютеров по партиям из задания
p_all = 500
p1 = 180
p2 = 170
p3 = p_all - (p1 + p2)
# задаем вероятности брака по партиям
b1 = 0.03
b2 = 0.02
b3 = 0.06

# P = round(((p1 / p_all) * b1 + (p2 / p_all) * b2 + (p3 / p_all) * b3), 4)
P = round((p1 * b1 + p2 * b2 + p3 * b3) / p_all, 4)
print(P)
```

Задание 2

Есть компания с программистами. В ней на языке Java пишут в 3 раза больше человек, чем на языке C++. Программистов на языке C++ в 4 раза больше, чем на Python. Вероятность сделать работу за один день для Java-программиста — 0,85, для программиста на языке C++ 0,9, а для Python программиста 0,8. Найти вероятность того, что программист, выбранный наугад, сделает работу за один день. Ответ округлите до сотых. В качестве десятичного разделителя используйте точку.

```python
# задаем количество программистов из задания
p = 1
c = p * 4
j = c * 3
total = j + p + c

# задаем вероятности выполнения работ за 1 день для разных типов программистов
jp = 0.85
cp = 0.9
pp = 0.8

P = round((p / total) * pp + (c / total) * cp + (j / total) * jp, 2)
print(P)
```

В ответ запишите наибольшую вероятность, округленную до тысячных. В качестве разделителя используйте точку.

Задание 4

Есть 4 кубика. На трех из них окрашена белым половина граней, а на четвертом кубике всего одна грань из шести белая. Наудачу выбранный кубик подбрасывается семь раз. Найти вероятность того, что был выбран четвертый кубик, если при семи подбрасываниях белая грань выпала ровно один раз.  
Ответ округлите до тысячных. Используйте точку в качестве десятичного разделителя.

```python
# задаем количество кубиков: 2 группы - 1 группа из 3-х одинаковых и 2 группа из одного
g1 = 3
g2 = 1
total = g1 + g2

# задаем вероятности выпадения белой грани для каждой группы
p_g1 = 0.5
p_g2 = 1 / 6

# задаем условия
n = 7 #количество вероятностей (7 бросков)
k = 1 # количество выпадений
c = special.comb(n, k) #биномиальный коэффициент
q_1 = 1 - p_g1 #вероятность выпадения черной грани для 1-ой группы
q_2 = 1 - p_g2 #вероятность выпадения черной грани для 2-ой группы
c = special.comb(n, k) #биномиальный коэффициент
#вероятность выпадения 1 белой грани при броске кубика из 1-ой и 2-ой групп
P_1 = c * (p_g1 ** k) * (q_1 ** (n - k))
P_2 = c * (p_g2 ** k) * (q_2 ** (n - k))

Pz = round(((g2 / total) * P_2) / (((g2 / total) * P_2) + ((g1 / total) * P_1)), 3)
Pz
```

На гитхабе лежит 12 скриптов, написанных аналитиком №1, 20 скриптов — написанных аналитиком №2 и 18 скриптов — написанных аналитиком №3. Вероятность того, что скрипт, написанный аналитиком №1, работает  без ошибок, равна 0,9; для скриптов, написанных аналитиками  №2 и №3, эти вероятности соответственно равны 0,6 и 0,9.

Нужно найти вероятность того, что случайно взятый скрипт будет работать без ошибок. В ответ запишите число, округленное до сотых. В качестве десятичного разделителя используйте точку.

```python
# т.к. вероятности аналитика 1 и аналитика 3 одинаковые
# и нам не требуется определять вероятность для конкретного аналитика
# мы можем объединить аналитика 1 и аналитика 3 в 1 группу (a1)
a1 = 12 + 18
a2 = 20
total = a1 + a2

# задаем вероятноси для каждой группы
p_a1 = 0.9
p_a2 = 0.6

P = round((a1 / total) * p_a1 + (a2 / total) * p_a2, 2)
P
```

Задание 7

Три фабрики производят лампочки. Фабрика А производит 20% всех лампочек, 50% лампочек производится на фабрике В и оставшиеся 30% на фабрике С. 2% лампочек, произведенных на фабрике А, 1% лампочек, произведенных на фабрике В, и 3% лампочек, произведенных на фабрике С, являются бракованными.На рынке наугад выбирается лампочка и оказывается бракованной. Какова вероятность того, что эта лампочка была произведена на заводе B?

Ответ округлите до тысячных, в качестве разделителя  используйте точку.

```python
# задаем % производства лампочек от общего количества
fa = 0.2
fb = 0.5
fc = 0.3
#  задаем вероятности брака для каждой фабрики
p_fa = 0.02
p_fb = 0.01
p_fc = 0.03

P_A = round((fa * p_fa) / ((fa * p_fa) + 
                                     (fb * p_fb) + 
                                     (fc * p_fc)), 3)
P_B = round((fb * p_fb) / ((fa * p_fa) + 
                                     (fb * p_fb) + 
                                     (fc * p_fc)), 3)
P_C = round((fc * p_fc) / ((fa * p_fa) + 
                                     (fb * p_fb) + 
                                     (fc * p_fc)), 3)
print(P_A, P_B, P_C)
```

Задание 8

Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета?

Ответ округлите до десятитысячных, в качестве десятичного разделителя используйте точку.

```python
p_obn = 0.98
pogr_obn = 1 - p_obn

p_lozh_srab = 0.05

p_est_sam = 0.07
p_net_sam = 1 - p_est_sam

P = (p_est_sam * pogr_obn) / ((p_est_sam * pogr_obn) + (p_net_sam * (1 - p_lozh_srab)))
round(P, 4)
```

Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. Ответ округлите до тысячных, в качестве десятичного разделителя используйте точку.
