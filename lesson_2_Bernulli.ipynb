Задание 1
Устройство, состоящее из пяти независимо работающих элементов, включается за время Т. Вероятность отказа каждого из них за это время равна 0,4. Найти вероятность того, что откажут три элемента. Ответ округлите до четырёх знаков после запятой. В качестве десятичного разделителя используйте точку.

```python
n = 5 #количество вероятностей (5 работающих элементов)
k = 3 # количество отказов
p = 0.4 #вероятность отказа
q = 1 - p #вероятность успешной работы
c = special.comb(n, k) #биномиальный коэффициент
P = round(c * (p ** k) * (q ** (n - k)), 4) #вероятность отказа + округление до 4 знака после точки

print(P)
```

Задание 4

В магазине 7 покупателей. Каждый может совершить покупку с вероятностью 0,4. Найти вероятность того, что не более двух человек совершат покупку. Ответ округлите до сотых. В качестве десятичного разделителя используйте точку.

```python
n = 7 #количество покупателей/вариантов
p = 0.4 #вероятность совершения покупки
q = 1 - p #вероятность отказа от покупки
k_0 = 0 # количество покупок (0)
k_1 = 1 # количество покупок(1)
k_2 = 2 # количество покупок (2)
c_0 = special.comb(n, k_0) #биномиальный коэффициент 0 покупок
c_1 = special.comb(n, k_1) #биномиальный коэффициент 1 покупка
c_2 = special.comb(n, k_2) #биномиальный коэффициент 2 покупки
P_0 = c_0 * (p ** k_0) * (q ** (n - k_0)) #вероятность, что купили 0 раз
P_1 = c_1 * (p ** k_1) * (q ** (n - k_1)) #вероятность, что купили 1 раз
P_2 = c_2 * (p ** k_2) * (q ** (n - k_2)) #вероятность, что купили 2 раза
print(round((P_0 + P_1 + P_2), 2)) #вывод на печать вероятности суммы вероятностей покупки + округление
```

Задание 7

Для нормальной работы отдела по дата-аналитике необходимо, чтобы в офисе было не менее восьми программистов, а всего в штате 10 программистов. Вероятность того, что программист заболеет, равна 0.1. Найдите вероятность нормальной работы отдела по дата-аналитике.

Ответ округлите до сотых, а в качестве разделителя используйте точку.

```python
n = 10 #количество программистов/вариантов
p = 0.1 #вероятность, что программист заболеет
q = 1 - p #вероятность того, что программист выйдет на работу
k_0 = 0 # количество программистов заболеет(0)
k_1 = 1 # количество программистов заболеет (1)
k_2 = 2 # количество программистов заболеет (2)
c_0 = special.comb(n, k_0) #биномиальный коэффициент 0 программистов заболеет
c_1 = special.comb(n, k_1) #биномиальный коэффициент 1 программист заболеет
c_2 = special.comb(n, k_2) #биномиальный коэффициент 2 программиста заболеет
P_0 = c_0 * (p ** k_0) * (q ** (n - k_0)) #вероятность, что 0 программистов заболеет
P_1 = c_1 * (p ** k_1) * (q ** (n - k_1)) #вероятность, что 1 программист заболеет
P_2 = c_2 * (p ** k_2) * (q ** (n - k_2)) #вероятность, что 2 программиста заболеет
#вывод на печать суммы вероятностей больных программистов (от 0 до 2) + округление
print(round((P_0 + P_1 + P_2), 2)) 
```

```python
n = 30 #количество скриптов/вариантов
p = 0.1 #вероятность, что скрипт с ошибкой
q = 1 - p #вероятность того, что скрипт без ошибки
k = 2 # количество скриптов с ошибкой(2)
c = special.comb(n, k) #биномиальный коэффициент 2 скрипта с ошибкой
P = c * (p ** k) * (q ** (n - k)) #вероятность, что 2 скрипта с ошибкой
#вывод на печать вероятности 2 скриптов с ошибкой + округление
print(round(P, 3)) 
```

Задание 10

В урне 20 шаров: 15 белых и 5 чёрных. Вынули подряд 5 шаров, причём каждый вынутый шар возвращается в урну и перед извлечением следующего шары в урне тщательно перемешиваются. Найдите вероятность того, что из пяти вынутых шаров будет 2 белых.

Ответ округлите до тысячных, в качестве десятичного разделителя используйте точку.

```python
n = 5 #количество вытягиванй/вариантов
p = 15 / 20 #вероятность вытянуть белый
q = 1 - p #вероятность вытянуть черный
k = 2 # количество вытянутых белых шаров(2)
c = special.comb(n, k) #биномиальный коэффициент 2 белых шара
P = c * (p ** k) * (q ** (n - k)) #вероятность, что вытянут 2 белых шара
#вывод вероятности, что вытянут 2 белых шара + округление
print(round(P, 3)) 
```
